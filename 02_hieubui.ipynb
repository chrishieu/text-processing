## Removal of Punctuations

```python
import string
PUNCT_TO_REMOVE = str(string.punctuation).replace("_", "")
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))
remove_punctuation("I ! ? . ;")
```

```python
def replace_underscore(x):
  return x.translate(x.maketrans('_', ' '))

x = 'NÆ¡i thÃ­ch_há»£p chgo há»™i nhÃ³m , háº¹n_hÃ² .\n'
replace_underscore(x)
```

# Remove urls & htmls

```python
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)
```

#Remove Digit

```python
def remove_digit(text):
    return text.translate(str.maketrans('', '', '0123456789'))

print(remove_digit('b4ngt4n1306'))
```

```python
import emot
emot_obj = emot.emot()
def convert_emoji(text):
  for emoji, mean in zip(emot_obj.emoji(text)['value'], emot_obj.emoji(text)['mean']):
    text = text.replace(emoji, mean.replace(":", ""))
  return text
convert_emoji("I love python â˜® ğŸ™‚ â¤ :-) :-( :-)))")
```

# Convert emoticon to text

```python
def convert_emoticons(text):
  dict_emoticons = dict(zip(emot_obj.emoticons(text)['value'], emot_obj.emoticons(text)['mean']))
  res_emoticons =  dict(sorted(dict_emoticons.items(), key = lambda kv:len(kv[1]), reverse=True))
  for emoticon, mean in res_emoticons.items():
    text = text.replace(emoticon, mean)
  return text
convert_emoticons("I love python â˜® ğŸ™‚ â¤ :-) :-( :-))) :-) :))")
```

## Remove rarewords

```python
from collections import Counter
counter = Counter()

def count_words(text):
  for word in text.split():
    counter[word] +=1
text = "hÃ  ná»™i sáº¯p mÆ°a hÃ  ná»™i Æ¡i"
count_words(text)
n_rare = 2
rare_words = set([w for w, wc in counter.most_common()[:-n_rare-1:-1]])
remove_rarewords = lambda text: " ".join([word for word in text.split() if word not in rare_words])
remove_rarewords("hÃ  ná»™i mÆ°a")
```

# Remove not alpha

```python
def remove_not_alpha(text):
  return ''.join(c for c in text if (c.isalpha()))
```

# Preprocess

```python
def preprocess(text):
  text = remove_urls(text)
  text = remove_html(text)
  text = remove_vni(text)
  text = convert_emoticons(text)
  text = convert_emoji(text)
  text = remove_punctuation(text)
  text = remove_digit(text)
  # text = remove_not_alpha(text)
  
  
  return text
```
